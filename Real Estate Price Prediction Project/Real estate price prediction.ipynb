```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (15,10)
```

```python
df = pd.read_csv('bengaluru_house_prices.csv')
df.head()
```

```python
df1 = df.drop(['area_type','availability','society','balcony'],axis = 'columns')
df1.head()
```

```python
df1.shape
```

```python
df1.isna().sum()
```

```python
df2 = df1.dropna()
```

```python
df2.isna().sum()
```

```python
df2['size'].unique()
```

```python
df2['rooms'] = df2['size'].apply(lambda x: int(x.split(' ')[0]))
df2.head()
```

```python
df3 = df2.drop('size',axis = 'columns')
df3.head()
```

```python
df3.shape
```

```python
df3['total_sqft'].unique()
```

```python
def sqft(x):
    y = x.split('-')
    if len(y)==2:
        return(float(y[0])+float(y[1]))/2
    try:
        return (float(x))
    except:
        return None
df3['total_sqft'] = df3['total_sqft'].apply(sqft)
```

```python
df3['total_sqft'].unique()
```

```python
df3['price'] = df3['price'].apply(lambda x: float(x))
df3['bath'] = df3['bath'].apply(lambda x: float(x))
df3.sample(10)
```

```python
df3['price_per_sqft'] = df3['price']*100000/df3['total_sqft']
df3.head()
```

```python
df3.location = df3.location.apply(lambda x: x.strip())
no_of_location = df3['location'].value_counts()
no_of_location
```

```python
len(no_of_location)
```

```python
location_under_10 = no_of_location[no_of_location<=10]
location_under_10.unique()
```

```python
df3.location = df3.location.apply(lambda x: 'other' if x in location_under_10 else x)
len(df3['location'].unique())
```

```python
df3.head()
```

```python
df4 = df3[df3['total_sqft']/df3['rooms']>=300]
df4.shape
```

```python
df4.describe()
```

```python
u = df4['price_per_sqft'].mean()+df4['price_per_sqft'].std()
l = df4['price_per_sqft'].mean()-df4['price_per_sqft'].std()
u,l
```
